```python
total_text = ''
for num in range(1,15):
    file = open(f"/content/drive/MyDrive/dataset/기호_{num} 공약.txt", 'r')
    while True:
        line = file.readline()
        if not line: break
        total_text += line
    file.close()
print(total_text)
f = open("/content/drive/MyDrive/dataset/result/모든공약.txt",'w')
f.write(total_text)
f.close()
```
